```python
### FACES ###
import numpy as np
import cv2
import pickle

# Import haar cascade for frontal face
face_cascade = cv2.CascadeClassifier('data/haarcascade_frontalface_alt2.xml')
# Add in recognizer
recognizer = cv2.face.LBPHFaceRecognizer_create()
# Add in train data
recognizer.read("trainer.yml")


"Load Label Names from Pickle"
# reverse the lookup
labels = {"person_name": 1}
with open("labels.pickle", 'rb') as f:  # 'wb' as I'm writing bytes, as file or 'f'
    # dump in label id to that file
    og_labels = pickle.load(f)
    # inverted actual label
    labels = {v:k for k,v in og_labels.items()}

cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    # Convert the frame to gray to let cascade works
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # find all the faces in this frame (video/picture)
    # if scaleFactor is increased a little bit, it might be more accurate but going too high is a problem
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.5, minNeighbors=5)

    # iterate through theses faces
    for (x, y, w, h) in faces:
        # region of interest (ROI) for gray = location of this frame
        roi_gray = gray[y:y+h, x:x+w]
        # we also can write ROI for color frame
        roi_color = frame[y:y+h, x:x+w]

        "Recognize ROI / person using deep learned model"
        # predict ROI => gives the id and how confidence it is
        id_, conf = recognizer.predict(roi_gray)
        if conf >= 45 and conf <= 85:
            
            print(id_)
            print(labels[id_])

            "Put Text"
            # Put text on the frame
            font = cv2.FONT_HERSHEY_SIMPLEX
            name = labels[id_]
            color = (255, 255, 255)
            stroke = 2
            cv2.putText(frame, name, (x,y), font, 1, color, stroke, cv2.LINE_AA)

        # save this image, can be used to rewrite your face in several times
        img_item = "my-image.png"
        cv2.imwrite(img_item, roi_color)

        "Draw a Rectangle"
        color = (255, 0, 0) # not in RGB but in BGR - not sure why
        # how thick we want the line to actually be
        stroke = 2
        end_cord_x = x + w
        end_cord_y = y + h
        # grab the frame, starting coordinates, how big it gonna be (wide, tall) _ ending coordinates
        cv2.rectangle(frame, (x, y), (end_cord_x, end_cord_y), color, stroke)
        

    # Display the resulting frame in color
    cv2.imshow('frame', frame)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break

# When everthing done, release the capture
cap.release()
cv2.destroyAllWindows()
```
